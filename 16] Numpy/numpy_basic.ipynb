```python
import numpy as np
a=np.array([1, 2, 3])
print(a)                    #[1 2 3]
print(type(a))


a1=np.array((1,2,3,4,5))
print(a1)                   #[1 2 3 4 5]
print(type(a1))


a2=np.array({1,2,3,4})
print(a2)                   #{1, 2, 3, 4}



a3=np.array({1:'a',2:'b'})
print(a3)                   #{1: 'a', 2: 'b'}



#Accessing  elements of array
print(a[0])
```

```python
import numpy as np
a=np.array([[1,2,3],[4,5,6]])       #to create matrix [[],[]]
print(a)

print(a.T)                      #Transpose

#scaler operations
s=np.ones((3,3))
print(s)
```
